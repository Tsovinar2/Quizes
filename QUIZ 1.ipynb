```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
titanic=pd.read_csv("titanic.csv")
titanic
```

####1) Արդյո՞ք երիտասարդները ավելի հաճախ են փրկվում:

```
a=len(titanic[titanic['Survived']==1][titanic['Age']<27])/len(titanic[titanic['Survived']==1])
print("\n", a,"\n")

#Որոշեցի 1 տողով գրել: Չնայած աշխատեցնելիս ինչ-որ թերություն է ցույց տալիս:
```

####2) Գտնել ուղևորների ողջ մնալու %-ները այն տղամարդկանց և կանանց մոտ, որոնք տոմսերը գնել են Fare-ի մեդիանից բարձր գնով:

####Կա՞ արդյոք տարբերություն:

```
a=titanic[titanic["Fare"] > np.median(titanic["Fare"])][titanic["Survived"]==1]
print("\n", len(titanic[titanic['Sex']=="female"][titanic["Fare"] > np.median(titanic["Fare"])][titanic["Survived"]==1])/len(a),"\n\n")

#Fare-ի մեդիանից բարձր գնով տոմս գնած փրկվողների մեջ կանանց տոկոսը:
```

```
print("\n", len(titanic[titanic['Sex']=="female"][titanic["Survived"]==1])/len(titanic[titanic["Survived"]==1]),"\n\n")

#Ընդհանուր փրկվողների մեջ կանանց տոկոսը:
```

####3)Ուղևորների տարիքների էմպիրիկ բաշխումը նկարել հիստոգրամի տեսքով:

```
female=titanic[titanic['Sex']=="female"]
male=titanic[titanic['Sex']=="male"]
female.hist(by="Sex", column="Age", bins=40, color='magenta', alpha=0.9, rwidth=0.8)
male.hist(by="Sex", column="Age", bins=40, color='dodgerblue', alpha=0.6, rwidth=0.8)

#titanic.hist(by="Sex", column="Age", bins=40, color='purple', alpha=0.6, rwidth=0.8) Սենց էդքան էլ սիրուն չի:
```

####4)Գտնել ողջ մնալու %-ները տարբեր կլասների (Pclass) միջև: Նկարել այն հիստագրամի տեսքով:

```

b=titanic.groupby('Pclass').agg({'Survived':'sum'})

rate=b['Survived']/len(titanic[titanic['Survived']==1])
b['Rate']=rate
b
```

```
Pclass_1=titanic[titanic['Pclass']==1]
Pclass_2=titanic[titanic['Pclass']==2]
Pclass_3=titanic[titanic['Pclass']==3]
Pclass_1.hist(by="Pclass", column="Survived", bins=10, color='lightcoral', alpha=0.6)
Pclass_2.hist(by="Pclass", column="Survived", bins=10, color='springgreen', alpha=0.6)
Pclass_3.hist(by="Pclass", column="Survived", bins=10, color='orchid', alpha=0.6)
```
